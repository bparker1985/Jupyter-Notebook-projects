```python
# This will drop some irrelevant coulumns, becasue we don't need to show everything.
car = car.drop(['Engine Fuel Type', 'Market Category', 'Vehicle Style', 'Popularity', 'Number of Doors', 'Vehicle Size'], axis=1)
car.head(5)
```

```python
# This is how to rename the columns to names that are more straight forward. 
# For example I don't need 'Engine HP', I can renmae it to just 'HP'.
car = car.rename(columns={'Engine HP':'HP', 'Engine Cylinders': 'Cylinders', 'Transmission Type': 'Transmission', 'Driven_Wheels': 'Drive Mode', 'highway MPG': 'MPG-H', 'city mpg': 'MPG-C', 'MSRP': 'Price'})
car.head(5)
```

```python
# This will the total number of rows and columns.
# 11914 rows & 10 columns.
car.shape
```

```python
# This shows rows containing duplicate data
duplicate_rows_car = car[car.duplicated()]
print('Number of duplicate rows: ', duplicate_rows_car.shape)
```

```python
# An important step in documentation and data cleaning is counting rows and columns before removing duplicates
car.count()
```

```python
# This will drop the duplicates
car = car.drop_duplicates()
car.count()
```

```python
# This will find the NULL values
print(car.isnull().sum())
```

```python
# This will drop the NULL Values
car = car.dropna()
car.count()
```

```python
# When I check for NULLs now there are none.
print(car.isnull().sum())
```

```python
# This will find the outliers in the data
sns.boxplot(x=car['Price'])
```

```python
# This will plot a histogram for the numbers of cars for each brand.
# Let's also add useful lables to the visualization
car.Make.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title('Number of cars by make')
plt.ylabel('Number of cars')
plt.xlabel('Make')
```
